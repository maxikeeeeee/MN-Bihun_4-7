```python
matrix = [[random.randint(1, 12) for _ in range(3)] for _ in range(3)]
matrix_np = np.array(matrix)

try:
    inverse_matrix = np.linalg.inv(matrix_np)
except np.linalg.LinAlgError:
    inverse_matrix = "Матриця не має оберненої."
```

```python
print("Матриця 3x3:")
print(matrix_np)
print("\nОбернена матриця:")
print(inverse_matrix)
```

```python
matrix = [[random.uniform(0, 1) for _ in range(4)] for _ in range(4)]

matrix_np = np.array(matrix)

row_products = np.prod(matrix_np, axis=1)

vector_column = row_products.reshape(-1, 1)
```

```python
print("Матриця 4x4:")
print(matrix_np)
print("\nВектор-стовпчик:")
print(vector_column)
```
