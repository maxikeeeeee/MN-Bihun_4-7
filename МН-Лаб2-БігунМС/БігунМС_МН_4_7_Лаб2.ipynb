```python
import pandas as pd
import numpy as np
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal)'
```

```python
df = pd.DataFrame({
    'Країна': ['Країна А', 'Країна Б', 'Країна В'],
    'MVF_2024': [3000, 1500, 2000],
    'WorldBank_2023': [2900, 1600, 2100],
    'OON_2022': [2800, 1700, 2200]
})
```

```python
data_first_5_rows = df.head()
```

```python
dataset_size = df.shape
```

```python
df.columns = ['Країна', 'MVF_2024', 'WorldBank_2023', 'OON_2022']
```

```python
df.dtypes
```

```python
df.replace('-', np.nan, inplace=True)
missing_values = df.isnull().sum()
```

```python
df.fillna(df.select_dtypes(include=[np.number]).mean(), inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
dataset_description = df.describe()
```

```python
df['Різниця_MVF_WB'] = df['MVF_2024'] - df['WorldBank_2023']
```

```python
df['Відсоток_Зміна_MVF_WB'] = ((df['MVF_2024'] - df['WorldBank_2023']) / df['WorldBank_2023']) * 100
df['Відсоток_Зміна_WB_OON'] = ((df['WorldBank_2023'] - df['OON_2022']) / df['OON_2022']) * 100
```

```python
країна_з_найбільшою_зміною = df.loc[df['Відсоток_Зміна_MVF_WB'].idxmax(), 'Країна']
```

```python
середнє_MVF_2024 = df['MVF_2024'].mean()
середнє_WorldBank_2023 = df['WorldBank_2023'].mean()
середнє_OON_2022 = df['OON_2022'].mean()
```

```python
df['Загальне_Коливання'] = df[['MVF_2024', 'WorldBank_2023', 'OON_2022']].std(axis=1)
країна_з_найбільшими_коливаннями = df.loc[df['Загальне_Коливання'].idxmax(), 'Країна']
```

```python
df['Внесок_MVF_2024'] = (df['MVF_2024'] / df['MVF_2024'].sum()) * 100
df['Внесок_WB_2023'] = (df['WorldBank_2023'] / df['WorldBank_2023'].sum()) * 100
df['Внесок_OON_2022'] = (df['OON_2022'] / df['OON_2022'].sum()) * 100
```

```python
import matplotlib.pyplot as plt
```

```python
df.set_index('Країна')[['MVF_2024', 'WorldBank_2023', 'OON_2022']].T.plot(kind='line')
plt.title('Зміни у ВВП для країн за три роки')
plt.xlabel('Рік')
plt.ylabel('ВВП (млн. доларів США)')
plt.show()
```

```python
print("Перші 5 рядків даних:")
print(data_first_5_rows)

print("\nРозмір набору даних (рядки, стовпці):")
print(dataset_size)

print("\nПропущені значення в даних:")
print(missing_values)

print("\nОпис набору даних:")
print(dataset_description)

print("\nРізниця між MVF_2024 та WorldBank_2023:")
print(df[['Країна', 'Різниця_MVF_WB']])

print("\nВідсоткова зміна між роками:")
print(df[['Країна', 'Відсоток_Зміна_MVF_WB', 'Відсоток_Зміна_WB_OON']])

print("\nКраїна з найбільшою відсотковою зміною:")
print(країна_з_найбільшою_зміною)

print("\nСередні значення для кожного показника:")
print(f"MVF_2024: {середнє_MVF_2024}")
print(f"WorldBank_2023: {середнє_WorldBank_2023}")
print(f"OON_2022: {середнє_OON_2022}")

print("\nКраїна з найбільшими коливаннями:")
print(країна_з_найбільшими_коливаннями)

print("\nВнесок кожної країни у загальну суму кожного показника:")
print(df[['Країна', 'Внесок_MVF_2024', 'Внесок_WB_2023', 'Внесок_OON_2022']])
```

**У цьому завданні ми проаналізували ВВП кількох країн за різні роки, обчислили їх різницю та відсоткові зміни між роками. Ми також визначили країни з найбільшою зміною і внесок кожної країни у загальний показник, що допомагає зрозуміти економічні тенденції.**
